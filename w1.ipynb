	- Terminal (ubuntu env)
	- xterm (simple)

	- gnome (gnome env - feature rich)

#### baby commands:
	- pwd 	(shows present working directory)
	- ls 	(shows list of folder & files in current directory)
	- ps 	(processes running in that particular shell)
	- uname (shows the OS)
	- clear OR ctrl+L
	- exit OR ctrl+D

>`adam@machine:~$ ls -a`
>-	`adam` 				= USERNAME
>-	`machine` 			= HOSTNAME
>-	`~` 				= PATH
>-	`adam@machine:~$` 	= COMMAND PROMPT
>-	`ls` 				= COMMAND
>-	`-a` 				= OPTION
---------------------------------
>`ls` command options
>-	`-a` = shows hidden file/directory
>-	`-l` = shows file/directory in long (detailed) format
---------------------------------
>-`man 1 ls`
>-	`man` = COMMAND
>-	`1`	= option
>-	`ls` = ARGUMENT

### W1.3 Simple Commands in Linux
>#### SIMPLE COMMANDS
>|Commands|Options|Remarks|
>|:----|:-----:|-----|
>|`date`| 	|shows timestamp|
>|	|`-R`| 	shows in RFC 2233 format|
>|`cal`|	 	|shows calender of a month|
>|	|`mmm yyyy`| 	e.g. aug 2002|
>|	|`mm yyyy`|		e.g. 8 2002|
>|`ncal`|	|shows calender in transposed formatting of "cal"|
>|`free`|	|shows the memory & swap status of device|
>|	|`-h`	|human readable format|
>|`groups`| 	|shows group(s) the user belongs to|

---------------------------------
>#### files 	(shows type of the file/directory)
> TYPICAL OUTPUT OF `ls -l`
>
>`drwxr-xr-x 5 adam adam 12288 Nov 13 12:00 Documents`
>-	 `d` 				=	FILE TYPE
>-	 `rwx`			= 	OWNER PERMISSIONS
>-	 `r-x(1)`			=	GROUP PERMISSIONS
>-	 `r-x(2)`			=	OTHER PERMISSIONS
>-	 `5` 				=	NO. OF HARD LINKS
>-	 `adam(1)` 		=	OWNER
>-	 `adam(2)`		=	GROUP
>-	 `12288` 			= 	SIZE
>-	 `Nov 13 12:00` 	= 	LAST MODIFIED TIMESTAMP
>-	 `Documents`		= 	FILENAME
---------------------------------
>FILE TYPES
>
>-	`-` 	= 	Regular file
>-	`d`	=	Directory (Folder)
>-	`l`	=	Symbolic link
>-	`c` 	=	Character file (usually Terminal)
>-	`b`	=	Block file (usually  Hard drive)
>-	`S`	=	Socket file 
>-	`p` 	= 	Named pipe
---------------------------------
>PERMISSION TYPES
>
>-	`-` 	=	no permission(0)
>-	`r` 	= 	read(4)
>-	`w` 	= 	write(2)
>-	`x` 	= 	executable(1)
---------------------------------
>PERMISSION STRING
>
>-	`---` 	0
>-	`--x` 	1(0+0+1)
>-	`r--`		4
>-	`r-x`		5
>-	`rw-`		6
>-	`rwx`		7(4+2+1)
>
>e.g. `rwxr-xr-x` means 755
---------------------------------
>PERMISSION MODIFICATION
>
>`chmod u+r filename`
>-	`u` 	user
>-	`g` 	group
>-	`o` 	other
>-	`+` 	grant permission
>-	`-` 	revoke permission
>
>`chmod 700 filename`
>- `700` i.e. `rwx------` (permission string as a number)
---------------------------------
>CREATE FILE/MODIFY TIMESTAMP
>
>`touch filename`
>	- changes the last modified timestamp of a file to current timestamp
>	- if file doesn't exist, it creates one
---------------------------------
>COPY FILE/DIRECTORY |  `cp old_file new_filename/new_file_path`
---------------------------------
>MOVE/RENAME FILE/DIRECTORY
    >- MOVE 	|  `mv old_file new_file_path`
    >- RENAME	|  `mv old_file new_filename`
---------------------------------
>REMOVE FILE/DIRECTORY |  `rm filename`
